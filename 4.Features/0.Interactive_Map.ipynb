```python
import pandas as pd
#from geopy.distance import geodesic
import folium
from folium import FeatureGroup, LayerControl, Map, Marker
#from sklearn import preprocessing
#import seaborn as sn
#import matplotlib.pyplot as plt
```

```python
hotels_df = pd.read_csv('Features_csv/dublin_hotels.csv')
tour_df = pd.read_csv('Features_csv/dublin_tourism.csv') 
uni_df = pd.read_csv('Features_csv/dublin_university.csv')
cafe_df = pd.read_csv('Features_csv/dublin_cafe.csv')
pub_df = pd.read_csv('Features_csv/dublin_pub.csv')
museum_df = pd.read_csv('Features_csv/dublin_museum.csv')
company_df = pd.read_csv('Features_csv/company_geo_list.csv')
rest_df = pd.read_csv('Features_csv/Restaurant_List_scope.csv')
```

```python
rest_df.sample(1)
```

# <a id="map">Interactive Map</a>

```python
dub_latitude = '53.341785'
dub_longitude = '-6.265288'
map_dublin = Map(location=[dub_latitude, dub_longitude], 
                        tiles='CartoDB dark_matter', zoom_start=13)
# defining layers  
rest_layer = FeatureGroup(name='Restaurant')
hotel_layer = FeatureGroup(name='Hotel')
tour_layer = FeatureGroup(name='Tourism')
uni_layer = FeatureGroup(name='University')
cafe_layer = FeatureGroup(name='Cafe')
pub_layer = FeatureGroup(name='Pub')
museum_layer = FeatureGroup(name='Museum')
company_layer = FeatureGroup(name='Company')


# hotel layer
for lat, lng, label in zip(hotels_df['Latitude'], hotels_df['Longitude'], hotels_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='red', fill=True,
            fill_color='red', fill_opacity=0.5, parse_html=False).add_to(hotel_layer)

# tour layer
for lat, lng, label in zip(tour_df['Latitude'], tour_df['Longitude'], tour_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='green', fill=True,
            fill_color='green', fill_opacity=0.5, parse_html=False).add_to(tour_layer)
    
# university layer
for lat, lng, label in zip(uni_df['Latitude'], uni_df['Longitude'], uni_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='purple', fill=True,
            fill_color='purple', fill_opacity=0.5, parse_html=False).add_to(uni_layer)
    
# cafe layer
for lat, lng, label in zip(cafe_df['Latitude'], cafe_df['Longitude'], cafe_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='brown', fill=True,
            fill_color='brown', fill_opacity=0.5, parse_html=False).add_to(cafe_layer)
    
# pub layer
for lat, lng, label in zip(pub_df['Latitude'], pub_df['Longitude'], pub_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='yellow', fill=True,
            fill_color='yellow', fill_opacity=0.5, parse_html=False).add_to(pub_layer)
    
# museum layer
for lat, lng, label in zip(museum_df['Latitude'], museum_df['Longitude'], museum_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='lightgreen', fill=True,
            fill_color='lightgreen', fill_opacity=0.5, parse_html=False).add_to(museum_layer)    
    
# company layer
for lat, lng, label in zip(company_df['Latitude'], company_df['Longitude'], company_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=2, popup=label, color='orange', fill=True,
            fill_color='orange', fill_opacity=0.5, parse_html=False).add_to(company_layer)
    
# restaurant layer
for lat, lng, label in zip(rest_df['Latitude'], rest_df['Longitude'], rest_df['Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=1, popup=label, color='blue', fill=True,
            fill_color='blue', fill_opacity=0.5, parse_html=False).add_to(rest_layer)
    
    
#adding layers to dublin map
hotel_layer.add_to(map_dublin)
tour_layer.add_to(map_dublin)
uni_layer.add_to(map_dublin)
cafe_layer.add_to(map_dublin)
pub_layer.add_to(map_dublin)
museum_layer.add_to(map_dublin)
company_layer.add_to(map_dublin)                                                                     
rest_layer.add_to(map_dublin)

LayerControl().add_to(map_dublin)

legend_html =   '''
                <div style="position: fixed; color: #00bfff; 
                            bottom: 30px; left: 30px; width: 135px; height: 265px; 
                            border:2px solid grey; z-index:9999; font-size:14px;
                            ">&nbsp; Dublin City Legend <br>
                              &nbsp; Restaurant &nbsp; <i class="fa fa-map-marker fa-2x" style="color:blue"></i><br>
                              &nbsp; Hotel &nbsp; <i class="fa fa-map-marker fa-2x" style="color:red"></i><br>
                              &nbsp; Tourism &nbsp; <i class="fa fa-map-marker fa-2x" style="color:green"></i><br>
                              &nbsp; University &nbsp; <i class="fa fa-map-marker fa-2x" style="color:purple"></i><br>
                              &nbsp; Cafe &nbsp; <i class="fa fa-map-marker fa-2x" style="color:brown"></i><br>
                              &nbsp; Museum &nbsp; <i class="fa fa-map-marker fa-2x" style="color:green"></i><br>
                              &nbsp; Company &nbsp; <i class="fa fa-map-marker fa-2x" style="color:orange"></i><br>
                              &nbsp; Pub &nbsp; <i class="fa fa-map-marker fa-2x" style="color:yellow"></i>
                </div>
                ''' 

map_dublin.get_root().html.add_child(folium.Element(legend_html))
map_dublin.save(outfile= "pics/map_features.html")

map_dublin
```

Use layer selection in order to activate/desactivate layers
