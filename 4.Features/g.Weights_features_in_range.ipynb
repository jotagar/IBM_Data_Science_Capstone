```python
import pandas as pd
from geopy.distance import geodesic
from sklearn import preprocessing
```

```python
hotels_df = pd.read_csv('Features_csv/dublin_hotels.csv')
tour_df = pd.read_csv('Features_csv/dublin_tourism.csv') 
uni_df = pd.read_csv('Features_csv/dublin_university.csv')
cafe_df = pd.read_csv('Features_csv/dublin_cafe.csv')
pub_df = pd.read_csv('Features_csv/dublin_pub.csv')
museum_df = pd.read_csv('Features_csv/dublin_museum.csv')
company_df = pd.read_csv('Features_csv/company_geo_list.csv')
rest_df = pd.read_csv('Features_csv/Restaurant_List_scope.csv')
```

# <a id="feat_gen">Feature Generation</a>

### Let's calculate the weights

We are taking into account 4 different ranges:
- < 250 meters
- 250 to 500 meters
- 500 to 1000 meters
- 1000 to 2000 meters

Once calculated the number of venues per range we multiply each one for a fixed number, which is proportional to the proximity to the restaurant

```python
space = '     ' # to print on same line while calculating
## subweighing: Once we get the number of the class within the range, 
# we count them and multiply for a fixed number: for the range 250 meters, it'd be 2000/250, and so on
w_s = [8, 4, 2, 1]
i=0
for i in range (0, len(rest_df)):
# hotel weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(hotels_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (hotels_df.at[j, 'Latitude'], hotels_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding hotel weight to a new column
    rest_df.at[i,'Hotel weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]

# tour weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(tour_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (tour_df.at[j, 'Latitude'], tour_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding tourism weight to a new column
    rest_df.at[i,'Tourism weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]

# university weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(uni_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (uni_df.at[j, 'Latitude'], uni_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding university weight to a new column
    rest_df.at[i,'University weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]
    
# cafe weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(cafe_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (cafe_df.at[j, 'Latitude'], cafe_df.at[j, 'Longitude'])).m
        
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding cafe weight to a new column
    rest_df.at[i,'Cafe weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]

# pub weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(pub_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (pub_df.at[j, 'Latitude'], pub_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding pub weight to a new column
    rest_df.at[i,'Pub weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]
    
# museum weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(museum_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (museum_df.at[j, 'Latitude'], museum_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding museum weight to a new column
    rest_df.at[i,'Museum weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]

# company weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(company_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (company_df.at[j, 'Latitude'], company_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding company weight to a new column
    rest_df.at[i,'Company weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]    
    
# restaurant weight
    d_250 = []
    d_500 = []
    d_1000 = []
    d_2000 = []    
    for j in range (0, len(rest_df)):
        d = geodesic((rest_df.at[i,'Latitude'], rest_df.at[i,'Longitude']), 
                        (rest_df.at[j, 'Latitude'], rest_df.at[j, 'Longitude'])).m
        if d > 0 and d <= 250:
            d_250.append(d)
        elif d > 250 and d <= 500:
            d_500.append(d)
        elif d > 500 and d <= 1000:
            d_1000.append(d)
        elif d > 1000 and d <= 2000:
            d_2000.append(d)
        else:
            continue
    # adding restaurant weight to a new column
    rest_df.at[i,'Restaurant weight'] = len(d_250)*w_s[0] + len(d_500)*w_s[1] + len(d_1000)*w_s[2] + len(d_2000)*w_s[3]    
    
#printing restaurant name to view the calculation progress
    print(rest_df.at[i,'Name'] +'{}'.format(10*space), end="\r", flush=True)
print('Completed.           ')
```

Normalizing weights

```python
weight_col = ['Hotel weight', 'Tourism weight', 'University weight', 
              'Cafe weight', 'Pub weight', 'Museum weight', 'Company weight', 'Restaurant weight']
weight_col_NOR = ['Hotel weight NOR', 'Tourism weight NOR', 'University weight NOR', 
                  'Cafe weight NOR', 'Pub weight NOR', 'Museum weight NOR', 'Company weight NOR', 'Restaurant weight NOR']
weight_nor = rest_df[weight_col].values
weight_nor = preprocessing.StandardScaler(with_mean=False, with_std=True).fit(weight_nor).transform(weight_nor.astype(float))
weight_nor_df = pd.DataFrame(data=weight_nor, columns=weight_col_NOR)
rest_df_NOR = pd.merge(rest_df, weight_nor_df, left_index=True, right_index=True )
rest_df_NOR.drop(columns=weight_col, inplace=True)
```
